```python
from matplotlib import pyplot as plt
from wordcloud import WordCloud

stop_word_freq = {
    'the': 138482,
    'and': 71442,
    'of': 68305,
    'a': 58789,
    'to': 57378,
    'i': 46576,
    'is': 40362,
    'in': 37107,
    'it': 34047,
    'this': 28968
}

# Using a default font that comes with matplotlib
wordcloud = WordCloud(width=800, height=400).generate_from_frequencies(stop_word_freq)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
